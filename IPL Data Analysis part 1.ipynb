### Batsman Analysis

```python
balls=df1.groupby(['batsman'])['ball'].count().reset_index()
balls
```

```python
Top_Batsman_balls=balls.nlargest(n=10, columns=['ball']).reset_index()
Top_Batsman_balls
```

```python
plt.figure(figsize=(12,8))
plt.title("Top Five batsman by ball played",fontsize=15,color='blue')
plt.xlabel("Balls")
plt.ylabel("Batsman")
plt.bar(Top_Batsman_balls['batsman'],Top_Batsman_balls['ball'],color='red')
```

```python
runs=df1.groupby(['batsman'])['batsman_runs'].sum().reset_index()
runs.columns=['batsman','runs']
runs
```

```python
Top_Batsman_by_runs=runs.nlargest(n=10, columns=['runs']).reset_index()
Top_Batsman_by_runs
```

```python
plt.figure(figsize=(12,8))
plt.title("Top Five batsman by runs",fontsize=15,color='blue')
plt.xlabel("Runs")
plt.ylabel("Batsman")
plt.bar(Top_Batsman_by_runs['batsman'],Top_Batsman_by_runs['runs'],color='cyan')
```

### Most Four & Sixes by batsman

```python
four=df1[df1['batsman_runs']==4]
four.head(3)
```

```python
runs_4=four.groupby('batsman')['batsman_runs'].count().reset_index()
runs_4.columns=['batsman','4s']
runs_4
```

```python
six=df1.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index()
six.columns=['batsman','6s']
six
```

```python
player=pd.concat([runs,balls.iloc[:,1],runs_4.iloc[:,1],six.iloc[:,1]],axis=1)
player
```

```python
player['Strike_rate']=player['runs']/player['ball']*100
player.head()
```

```python
player.isna().sum()
```

```python
player['4s'].fillna(0,inplace=True)
player.isnull().values.any()
```

```python
max_strike_rate=player.groupby(['batsman'])['Strike_rate'].sum()
plt.title("Top Five Batsman by strike rate")
max_strike_rate.sort_values(ascending=False)[:5].plot.bar(color='green')
```

```python
max_4s=player.groupby(['batsman'])['4s'].sum()
plt.title("Top Five Batsman by Fours hit")
max_4s.sort_values(ascending=False)[:5].plot.bar(color='orange')
```

```python
max_6s=player.groupby(['batsman'])['6s'].sum()
plt.title("Top Five Batsman by Sixes hit")
max_6s.sort_values(ascending=False)[:5].plot.bar(color='purple')
```
